```python
#Q1. Write a program to find all pairs of an integer array whose sum is equal to a given number?
class PairFinder:
    def __init__(self, arr):
        self.arr = arr
        self.seen = set()

    def find_pairs_with_sum(self, target_sum):
        pairs = []

        for num in self.arr:
            complement = target_sum - num
            if complement in self.seen:
                pairs.append((num, complement))
            self.seen.add(num)

        return pairs

# Example usage
arr = [2, 4, 3, 5, 6, -2, 8, 9]
target_sum = 7

pair_finder = PairFinder(arr)
result = pair_finder.find_pairs_with_sum(target_sum)

if result:
    print(f"Pairs with sum {target_sum}:")
    for pair in result:
        print(pair)
else:
    print(f"No pairs found with sum {target_sum}")
```

```python
 #Q2.Write a program to reverse an array in place? In place means you cannot create a new array. You have to update the original array.
class ArrayReverser:
    @classmethod
    def reverse_array_in_place(cls, arr):
        start = 0
        end = len(arr) - 1

        while start < end:
            arr[start], arr[end] = arr[end], arr[start]
            start += 1
            end -= 1

# Example usage
arr = [1, 2, 3, 4, 5]
print("Original array:", arr)

ArrayReverser.reverse_array_in_place(arr)
print("Reversed array:", arr)
```

```python
#Q3. Write a program to check if two strings are a rotation of each other?
def are_rotations(str1, str2):
    if len(str1) != len(str2):
        return False

    temp = str1 + str1
    if str2 in temp:
        return True

    return False

# Example usage
string1 = "abcdef"
string2 = "defabc"

if are_rotations(string1, string2):
    print(f"'{string1}' and '{string2}' are rotations of each other.")
else:
    print(f"'{string1}' and '{string2}' are not rotations of each other.")
```

```python
#Q4. Write a program to print the first non-repeated character from a string?
def first_non_repeated_char(string):
    char_count = {}

    for char in string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    for char in string:
        if char_count[char] == 1:
            return char

    return None

# Example usage
input_string = "abacdbe"
result = first_non_repeated_char(input_string)

if result:
    print(f"The first non-repeated character in '{input_string}' is '{result}'.")
else:
    print(f"There is no non-repeated character in '{input_string}'.")
```

```python
#Q5. Read about the Tower of Hanoi algorithm. Write a program to implement it
def tower_of_hanoi(n, source, auxiliary, target):
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return
    tower_of_hanoi(n - 1, source, target, auxiliary)
    print(f"Move disk {n} from {source} to {target}")
    tower_of_hanoi(n - 1, auxiliary, source, target)

# Example usage
num_disks = 3
tower_of_hanoi(num_disks, 'A', 'B', 'C')
```

```python
#Q6. Read about infix, prefix, and postfix expressions. Write a program to convert postfix to prefix expression.
def postfix_to_prefix(postfix_expr):
    stack = []
    
    for token in postfix_expr:
        if token.isalnum():  # Operand
            stack.append(token)
        else:  # Operator
            operand2 = stack.pop()
            operand1 = stack.pop()
            new_expr = token + operand1 + operand2
            stack.append(new_expr)
    
    return stack[0]

# Example usage
postfix_expression = "23+5*"
prefix_expression = postfix_to_prefix(postfix_expression)
print("Postfix Expression:", postfix_expression)
print("Prefix Expression:", prefix_expression)

```

```python
#Q7. Write a program to convert prefix expression to infix expression.
class PrefixToInfixConverter:
    def __init__(self):
        self.stack = []

    def is_operator(self, char):
        return char in ('+', '-', '*', '/', '^')

    def prefix_to_infix(self, prefix_expr):
        for token in reversed(prefix_expr):
            if token.isalnum():  # Operand
                self.stack.append(token)
            elif self.is_operator(token):  # Operator
                operand1 = self.stack.pop()
                operand2 = self.stack.pop()
                new_expr = f"({operand1}{token}{operand2})"
                self.stack.append(new_expr)

        return self.stack[0]

# Example usage
prefix_expression = "*+23*5+45"
converter = PrefixToInfixConverter()
infix_expression = converter.prefix_to_infix(prefix_expression)
print("Prefix Expression:", prefix_expression)
print("Infix Expression:", infix_expression)
```

```python
#Q8. Write a program to check if all the brackets are closed in a given code snippet.
class BracketChecker:
    def __init__(self):
        self.stack = []
        self.bracket_map = {')': '(', ']': '[', '}': '{'}

    def are_brackets_closed(self, code):
        for char in code:
            if char in '([{':
                self.stack.append(char)
            elif char in ')]}':
                if not self.stack or self.stack[-1] != self.bracket_map[char]:
                    return False
                self.stack.pop()

        return len(self.stack) == 0

# Example usage
code_snippet = """
def are_brackets_closed(code):
    stack = []
    for char in code:
        if char == '(':
            stack.append(char)
        elif char == ')':
            if not stack or stack[-1] != '(':
                return False
            stack.pop()
    return len(stack) == 0
"""

checker = BracketChecker()

if checker.are_brackets_closed(code_snippet):
    print("All brackets are properly closed.")
else:
    print("Not all brackets are properly closed.")
```

```python
#Q9. Write a program to reverse a stack.
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

    def reverse(self):
        if not self.is_empty():
            item = self.pop()
            self.reverse()
            self.insert_at_bottom(item)

    def insert_at_bottom(self, item):
        if self.is_empty():
            self.push(item)
        else:
            temp = self.pop()
            self.insert_at_bottom(item)
            self.push(temp)

    def display(self):
        print(self.items)

# Example usage
stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
print("Original Stack:")
stack.display()

stack.reverse()
print("\nReversed Stack:")
stack.display()
```

```python
#Q10. Write a program to find the smallest number using a stack.
class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []

    def push(self, item):
        self.stack.append(item)
        if not self.min_stack or item <= self.min_stack[-1]:
            self.min_stack.append(item)

    def pop(self):
        if not self.is_empty():
            if self.stack[-1] == self.min_stack[-1]:
                self.min_stack.pop()
            return self.stack.pop()

    def is_empty(self):
        return len(self.stack) == 0

    def get_min(self):
        if not self.is_empty():
            return self.min_stack[-1]

    def display(self):
        print(self.stack)

# Example usage
stack = MinStack()
stack.push(3)
stack.push(5)
stack.push(2)
stack.push(8)

print("Original Stack:")
stack.display()
print("Smallest number in stack:", stack.get_min())

stack.pop()
stack.pop()

print("\nUpdated Stack:")
stack.display()
print("Smallest number in stack:", stack.get_min())
```
